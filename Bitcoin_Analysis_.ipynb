```python
#Performing Exploratory Data Analysis (EDA)
# Displaying basic statistics of the dataset
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# Plotting the Close price over time
plt.figure(figsize=(10, 5))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Bitcoin Closing Price Over Time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
# Plotting the distribution of Bitcoin trading volume
plt.figure(figsize=(8, 5))
plt.hist(df['Volume'], bins=10, color='green', alpha=0.7)
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.title('Distribution of Bitcoin Trading Volume')
plt.show()
```

```python
import seaborn as sns

# Creating pair plot
sns.pairplot(df[['Open', 'Close', 'High', 'Low']])
plt.show()
```

```python
# Calculating the correlation between the columns
correlation_matrix = df[['Open', 'High', 'Low', 'Close', 'Volume']].corr()
print(correlation_matrix)
```

```python
# Filtering DataFrame for rows where Close price is greater than Open price
positive_gain_df = df[df['Close'] > df['Open']]
print(positive_gain_df)
```

```python
# Calculating the average volume
average_volume = df['Volume'].mean()

# Filtering DataFrame where trading volume is higher than the average
high_volume_df = df[df['Volume'] > average_volume]
print(high_volume_df)
```

```python
# Adding a new column 'Price Change' (difference between Close and Open)
df['Price Change'] = df['Close'] - df['Open']
print(df.head())
```

```python
# Apply if-else to create the 'Price Trend' column
df['Price Trend'] = df.apply(lambda row: 'Up' if row['Close'] > row['Open'] else 'Down', axis=1)
print(df[['Date', 'Open', 'Close', 'Price Trend']].head())
```
